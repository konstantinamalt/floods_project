```python
import pandas as pd
```

```python
df = pd.read_excel("CLEAN_FINAL_aois_all_details_nuts_translated_WITH_DEATHS_AND_STORM_NAME.xlsx")
```

```python
storm_boris = df[df['storm_name'] == 'Storm Boris']
storm_boris
```

```python
storm_boris_hectares = storm_boris[storm_boris['unit_measurement'] == 'hectares']
storm_boris_hectares
```

```python
storm_boris_hectares_extent = storm_boris_hectares[storm_boris_hectares['normalized_category_v2'] == 'event extent']
storm_boris_hectares_extent['affected'].sum()
```

```python
storm_boris_land = storm_boris_hectares[storm_boris_hectares['category_unit'] == 'Land use']
```

```python
storm_boris_land.groupby('normalized_category_v2')['affected'].sum().sort_values(ascending=False)
```

```python
storm_boris_land = storm_boris_land[storm_boris_land['normalized_category_v2'] != 'other land']
storm_boris_land
```

```python
countries_affected_all_land_BORIS = (
    storm_boris_land
    .groupby(["countries", "normalized_category_v1", "normalized_category_v2"])['affected']
    .sum()
    #.sort_values(ascending=False)
    .reset_index()
)
countries_affected_all_land_BORIS
```

```python
countries_affected_all_land_BORIS.to_excel("countries_affected_all_land_BORIS.xlsx", index=False, float_format="%.2f")
```

```python
storm_boris_km = storm_boris[storm_boris['unit_measurement'] == 'kilometres']
storm_boris_km_trans = storm_boris_km[storm_boris_km['category_unit'] == 'Transportation infrastructure']
storm_boris_km_trans = storm_boris_km_trans[storm_boris_km_trans['normalized_category_v2'] != 'other infrastructure']
```

```python
affected_transportation_types_details = (
    storm_boris_km_trans
    .groupby('normalized_category_v1')['affected']
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)

affected_transportation_types_details
```

```python
affected_transportation_types_details = (
    storm_boris_km_trans
    .groupby('normalized_category_v2')['affected']
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)

affected_transportation_types_details
```

```python
storm_boris_km_trans_new = storm_boris_km_trans[storm_boris_km_trans['normalized_category_v2'] != 'cart tracks']
```

```python
countries_affected_all_trans = (
    storm_boris_km_trans_new
    .groupby(["countries", "normalized_category_v2"])['affected']
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)
countries_affected_all_trans
```

```python
storm_boris_pop = storm_boris[storm_boris['normalized_category_v2'] == 'population']
```

```python
countries_affected_pop = (
    storm_boris_pop
    .groupby("countries")['affected']
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)
countries_affected_pop
```

## pipelines and communication

```python
storm_boris_facilities_km = storm_boris[storm_boris['unit_measurement'] == 'kilometres']
storm_boris_facilities_km
```

```python
storm_boris_pipelines_communication = storm_boris_facilities_km[storm_boris_facilities_km['normalized_category_v2'] == 'pipelines and communication infrastructure']
```
