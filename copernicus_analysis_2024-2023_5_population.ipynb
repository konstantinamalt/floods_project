```python
import pandas as pd
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', None)
```

```python
df = pd.read_excel("CLEAN_FINAL_aois_all_details_nuts_translated.xlsx")
```

```python
df_pop = df[df['normalized_category_v2'] == 'population']
```

```python
df_pop['affected'].sum()
```

```python
df['eventTime'] = pd.to_datetime(df['eventTime'])
df_pop = df[df['normalized_category_v2'] == 'population']
df_pop['year'] = df_pop['eventTime'].dt.year
df_2024 = df_pop[df_pop['year'] == 2024]
```

```python
df_2024['affected'].sum()
```

```python
countries_affected_pop = (
    df_pop
    .groupby("countries")['affected']
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)
countries_affected_pop
```

```python
countries_affected_pop.to_excel("countries_affected_population.xlsx", index=False, float_format="%.2f")
```

```python
nuts3_affected_pop = (
    df_pop
    .groupby(["nuts3_name_2024_en", "nuts3_2024_urban_rural_typology", "nuts3_2024_coastal_regions"])['affected']
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)
nuts3_affected_pop
```

```python
nuts3_affected_pop.to_excel("nuts3_affected_pop.xlsx", index=False, float_format="%.2f")
```

```python
nuts3_affected_pop_without_valencia = nuts3_affected_pop[
    nuts3_affected_pop['nuts3_name_2024_en'].str.strip() != 'Valencia province'
]
```

```python
nuts3_affected_pop.groupby("nuts3_2024_urban_rural_typology")['affected'].sum()
```
